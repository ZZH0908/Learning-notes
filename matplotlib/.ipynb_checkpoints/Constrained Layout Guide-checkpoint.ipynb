> constrained_layout会自动调整子图和像egends和colorbars这样的装饰位置，使布局更加合理
合理布局的前提是参数constrained_layout=True或plt.rcParams['figure.constrained_layout.use'] = True

### 例子

```python
import warnings
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
import matplotlib.gridspec as gridspec

import matplotlib._layoutbox as layoutbox

plt.rcParams['savefig.facecolor'] = "0.5"
plt.rcParams['figure.figsize'] = 4.5, 4.


def example_plot(ax, fontsize=12, nodec=False):
    ax.plot([1, 2])

    ax.locator_params(nbins=3)
    if not nodec:
        ax.set_xlabel('x-label', fontsize=fontsize)
        ax.set_ylabel('y-label', fontsize=fontsize)
        ax.set_title('Title', fontsize=fontsize)
    else:
        ax.set_xticklabels('')
        ax.set_yticklabels('')


# fig, ax = plt.subplots(constrained_layout=False)
# example_plot(ax, fontsize=24)

# fig, ax = plt.subplots(constrained_layout=True)
# example_plot(ax, fontsize=24)

fig, axs = plt.subplots(2, 2, constrained_layout=False) # 出现重叠，很不美观而且部分信息无法辨认
for ax in axs.flatten():
    example_plot(ax)
plt.show()
```

### Colorbars

```python
arr = np.arange(100).reshape((10, 10))
norm = mcolors.Normalize(vmin=0., vmax=100.)
pc_kwargs = {'rasterized': True, 'cmap': 'viridis', 'norm': norm}
fig, axs = plt.subplots(2, 2, figsize=(4, 4), constrained_layout=True)
for ax in axs.flatten():
    im = ax.pcolormesh(arr, **pc_kwargs) # pcolormesh()创建不规则的伪彩色图
    fig.colorbar(im, ax=ax, shrink=0.6) # ax参数会给给定轴移出空间来防止colorbar
# fig.colorbar(im, ax=axs, shrink=0.6)

plt.show()
```

```python
# 还可以使用切片的方式来对多子图的colorbar腾出空间并放置
fig, axs = plt.subplots(3, 3, figsize=(4, 4), constrained_layout=True)
for ax in axs.flatten():
    im = ax.pcolormesh(arr, **pc_kwargs)
fig.colorbar(im, ax=axs[1:, ][:, 1], shrink=0.8)
fig.colorbar(im, ax=axs[:, -1], shrink=0.6)
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(4, 4), constrained_layout=True)
for ax in axs[:2]:
    im = ax.pcolormesh(arr, **pc_kwargs)
fig.colorbar(im, ax=axs[:2], shrink=0.6)
im = axs[2].pcolormesh(arr, **pc_kwargs)
fig.colorbar(im, ax=axs[2], shrink=0.6)
# fig.colorbar(im, ax=[axs[2]], shrink=0.6) # 对单个轴以列表的形式传入，作用是使得colorbar同等对待，布局更美观合理
plt.show()
```

### 填充和间隔

```python
# 通过set_constrained_layout_pads进行设置，pad是指子图和图边缘的填充，space是指相邻图之间的间隔
# 在jupyter中效果不明显，建议在非交互模式下进行，即IDE或文本编辑运行
fig, axs = plt.subplots(2, 2, constrained_layout=True)
for ax in axs.flatten():
    example_plot(ax, nodec=True)
    ax.set_xticklabels('')
    ax.set_yticklabels('')
fig.set_constrained_layout_pads(w_pad=25./72., h_pad=25./72.,
        hspace=0., wspace=0.)

fig, axs = plt.subplots(2, 2, constrained_layout=True)
for ax in axs.flatten():
    example_plot(ax, nodec=True)
    ax.set_xticklabels('')
    ax.set_yticklabels('')
fig.set_constrained_layout_pads(w_pad=2./72., h_pad=2./72.,
        hspace=0., wspace=0.)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, constrained_layout=True)
for ax in axs.flatten():
    pc = ax.pcolormesh(arr, **pc_kwargs)
    fig.colorbar(pc, ax=ax, shrink=0.6, pad=0.2) # pad对colorbar与图之间进行填充
    ax.set_xticklabels('')
    ax.set_yticklabels('')
fig.set_constrained_layout_pads(w_pad=2./72., h_pad=2./72.,
        hspace=0.2, wspace=0.2)
plt.show()
```

### 手动设置axes的位置

```python
fig, axs = plt.subplots(1, 2)
example_plot(axs[0], fontsize=12)
axs[1].set_position([0.2, 0.2, 0.4, 0.4])
plt.show()
```

### constrained_layout的限制
constrained_layout参数是使用一种嵌套的方式来执行layout的

```python
# 在subplot中是无效的
# 原因: The reason is that each of these commands creates a separate GridSpec instance 
# and constrained_layout uses (nested) gridspecs to carry out the layout.
# 大意是每个subplot都会创建单独的GridSpec(网格规则)实例
fig = plt.figure(constrained_layout=True)

ax1 = plt.subplot(221)
print(type(ax1))
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
plt.show()
```

```python
# 对一个gridspec是可行的
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(2, 2)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[:, 1])

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
plt.show()
```

```python
# 同理也是不行的，每一行命令都创建一个不同的父gridspec，gridspec翻译为网格规则
fig = plt.figure(constrained_layout=True)

ax1 = plt.subplot2grid((3, 3), (0, 0))
ax2 = plt.subplot2grid((3, 3), (0, 1), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 0), colspan=2, rowspan=2)
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)
plt.show()
```

总结：subplot2grid和subplot都返回一个单独的不同的网格实例，无法统一协调；但constrained_layout参数定义是使用一种嵌套的方式来执行layout的，像fig.add_gridspec以及subplots都是这种方式，即先返回一个整个布局再往下分配，而不是单独创建再拼车~~
