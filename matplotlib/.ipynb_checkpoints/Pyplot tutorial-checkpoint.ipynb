### Intro to pyplot

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4]) # 如果只传入一个列表序列，则默认为Y值，x轴值是自动创建的；python的默认起始值为0；
plt.ylabel('some numbers')
plt.show()
```

```python
# 还可以这样传入数据
# 输入虽然是列表，但在内部已经间接的转换为np array
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 第三个参数是格式化字符串指示着颜色和线条类型
plt.axis([0, 6, 0, 20]) # [xmin, xmax, ymin, ymax]
plt.show()
```

```python
import numpy as np

t = np.arange(0., 6., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### Plotting with keyword strings

```python
# matplotlib允许使用带有data关键字的对象实例
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) # 前三个参数依次为x、y数据、颜色、点的大小，**kwargs 
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Plotting with categorical variables

```python
import matplotlib.pyplot as plt
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### 线条的属性

```python
import numpy as np
x = np.linspace(0, 2, 100)
# plt.plot(x, x, linewidth=4.0) # 设置线条宽度

# line, = plt.plot(x, x**2, '-') # 返回若干实例对象,以元组的形式
# print(type(line)) # 返回matplotlib.lines.Line2D的类实例,然后使用类方法来进行属性设置
# line.set_antialiased(True) # 是否开启反锯齿

lines = plt.plot(x, x, x, x**2)
plt.setp(lines, color='r', linewidth=2.0) # 若返回多条线,使用setp()函数进行统一设置会更为方便一些
plt.show()
```

```python
# 获得line属性
lines = plt.plot([1, 2, 3])
plt.setp(lines) # 查看可设置的属性
# plt.getp(lines) # 查看该line或lines的属性
```

### 多窗口与多axes操作

```python
# 例子
plt.figure(1)          
plt.subplot(211) # 在当前窗口add子窗口            
plt.plot([1, 2, 3])
plt.subplot(212)           
plt.plot([4, 5, 6])


plt.figure(2)               
plt.plot([4, 5, 6])       

plt.figure(1)                
plt.subplot(211)             
plt.title('Easy as 1, 2, 3') 
plt.show()
```

### 使用文本

```python
import numpy as np
import matplotlib.pyplot as plt
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') # latex公式编辑，形式：r'$.....$'；前缀r的作用是抑制转意符号；
# 以上的返回值都是<class 'matplotlib.text.Text'>的实例
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
# 参数为三维的或一个三位的整数，返回值没有图实例只有轴
# 比plt.subplots少个s，函数作用是Add a subplot to the current figure.
ax = plt.subplot(111) 
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             ) # 用文本来注释一个点；参数依次是注释文本、要注释的点、文本所在位置、箭头工具

plt.ylim(-2, 2)
plt.show()
```
