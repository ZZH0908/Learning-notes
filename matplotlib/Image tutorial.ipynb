```python
img = mpimg.imread('./images/czym.jpg')/255
img1 = img[:, :, 0]
plt.hist(img1.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
plt.show()
```

```python
imgplot = plt.imshow(img1, clim=(0.0, 0.7)) # 设置通道的颜色范围
```

```python
from PIL import Image

img = Image.open('./images/test.jpg')
# print(img.size)
img.thumbnail((30, 30), Image.ANTIALIAS)  # thumbnail只能缩小，不能放大
# print(img.size)
# plt.show() 在交互式模式中无效、没有作用
imgplot = plt.imshow(img)
```
