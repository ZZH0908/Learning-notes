### 组成：
1、多个条目，每个条目都是由label和key组成；
2、key相当于一个彩色标记
3、label是描述key的文本
4、legend handle
核心函数legend()

### Controlling the legend entries

```python
# 若不传入参数，则自动获取legend handles和相关label
import numpy as np
import matplotlib.pyplot as plt

ax = plt.gca()
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
plt.show() # 当前图和轴没有任何信息，则获取的handles、labels也是空的
```

```python
# line_up, = plt.plot([1,2,3], label='Line 2')
# line_down, = plt.plot([3,2,1], label='Line 1')
# plt.legend(handles=[line_up, line_down])

# 有时候忘了设置label，也可向legend传入一个labels list进行批量设置
line_up, = plt.plot([1,2,3])
line_down, = plt.plot([3,2,1])
plt.legend([line_up, line_down], ['Line Up', 'Line Down'])
plt.show()
```

```python
# 有时候不是所有handle都可以转换为条目，比如不是像上图的线条或者像第一个图压根没有具体的handle
# 所以要手动设置
import matplotlib.patches as mpatches
red_patch = mpatches.Patch(color='red', label='The red data')
plt.legend(handles=[red_patch])
plt.show()
# 还有很多标记，不仅仅是只有这样单调的patch
```

### Legend位置

```python
red_patch = mpatches.Patch(color='red', label='The red data')
# 第二个参数为位置设置，若为两个元素的元组即设计坐标，若为4个元素多的元组参数即设置ledend的长宽
# 第三个参数个人理解为ledend位置改变是相对于Axes还是Figure的，bbox_transform=plt.gca().transAxes则是相对于整个轴区域的
plt.legend(handles=[red_patch], bbox_to_anchor=(1, 1),
           bbox_transform=plt.gcf().transFigure)
plt.show()
```

```python
plt.subplot(211)
plt.plot([1, 2, 3], label="test1")
plt.plot([3, 2, 1], label="test2")
# ncol表示一个legend条目的列数
# expand表示legend按水平方向扩展至极限，None时则遵循bbox_to_anchor的size的设置
# borderaxespad表示legend和axes之间是否要pad，即设置二者之间的距离
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)

plt.subplot(223)
plt.plot([1, 2, 3], label="test1")
plt.plot([3, 2, 1], label="test2")
# Place a legend to the right of this smaller subplot.
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=2.)

plt.show()
```

### 一图多legend

```python
line1, = plt.plot([1, 2, 3], label="Line 1", linestyle='--')
line2, = plt.plot([3, 2, 1], label="Line 2", linewidth=4)

# Create a legend for the first line.
first_legend = plt.legend(handles=[line1], loc='upper right')

# Add the legend manually to the current Axes.
ax = plt.gca().add_artist(first_legend)

# Create another legend for the second line.
plt.legend(handles=[line2], loc='lower right')

plt.show()
```
