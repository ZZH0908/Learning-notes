> Matplotlib具有广泛的文本支持，包括对数学表达式的支持、对raster 和向量输出的truetype支持、具有任意旋转的换行分隔文本和unicode支持;用户对文本属性(字体大小、字体重量、文本位置和颜色等)有很大的控制，rc文件中设置了合理的默认值.另外，Matplotlib还支持大量LaTeX数学符号和命令

创建文本的命令函数有：text、annotate、xlabel、ylabel、title、figtext、subtitle，这些函数都返回一个text实例

```python
import matplotlib
import matplotlib.pyplot as plt

fig = plt.figure()
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
ax.set_title('axes title')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')

ax.text(3, 8, 'boxed italics text in data coords', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})

ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)

ax.text(3, 2, 'unicode: Institut für Festkörperphysik')
# transform=ax.transAxes表明文本是相对于轴放置的
ax.text(0.95, 0.01, 'colored text in axes coords',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='green', fontsize=15)


ax.plot([2], [1], 'o')
ax.annotate('annotate', xy=(2, 1), xytext=(3, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.axis([0, 10, 0, 10])

plt.show()
```

### 标题、标签

```python
# 设置字体属性
from matplotlib.font_manager import FontProperties
import numpy as np

x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
font.set_style('italic')

fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1)
# 对label进行位置设置,且label可以多行
ax.set_xlabel('time [s]\n This was a long experiment', fontproperties=font, position=(0., 1e6), horizontalalignment='left', fontsize='large', fontweight='bold')
ax.set_ylabel('Damped oscillation [V]')
plt.show()
# 补充：
# fig.subplots_adjust(top=0.8)设置顶部里边缘的距离
# ax.set_title('Vertically offset title', pad=30) 设置label与子图直接的填充
```

```python
# 公式作为轴label
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.2, left=0.2)
ax.plot(x1, np.cumsum(y1**2))
ax.set_xlabel('time [s] \n This was a long experiment')
ax.set_ylabel(r'$\int\ Y^2\ dt\ \ [V^2 s]$')
plt.show()
```

### 刻度和刻度标签

在制作图表时，放置刻度和ticktags是一个非常棘手的方面。Matplotlib自动地做到了最好，但是它也提供了一个非常灵活的框架来决定标记位置的选择以及如何标记它们；一个Axes对象包含有主刻度和次刻度，matplotlib.xaxis.set_major_locator和matplotlib.xaxis.set_minor_locator对应刻度位置绘制设定，matplotlib.xaxis.set_major_formatter和matplotlib.xaxis.set_minor_formatters对应刻度标签(数字)的格式.

#### Simple ticks

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
axs[1].xaxis.set_ticks(np.arange(0., 8.1, 2.))
plt.show() # 不美观，并且第二个图空白部分多余
```

```python
# 解决办法，更改刻度
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)

ticks = np.arange(0., 8.1, 2.)
# locator = matplotlib.ticker.FixedLocator(ticks) # 可以理解为封装的自适应方法，可以代替上一行代码

# 设置刻度值格式
tickla = ['%1.2f' % tick for tick in ticks] 
# formatter = matplotlib.ticker.StrMethodFormatter('{x:1.1f}') # 另一种设置格式的方法

axs[1].xaxis.set_ticks(ticks)
axs[1].xaxis.set_ticklabels(tickla)
axs[1].set_xlim(axs[0].get_xlim())
print(axs[0].get_xlim()) # 更改刻度范围限制
plt.show()
```

```python
# 默认的formatter为matplotlib.ticker.MaxNLocator
fig, axs = plt.subplots(2, 2, figsize=(8, 5), tight_layout=True)
axs = axs.flatten()
for n, ax in enumerate(axs):
    ax.plot(x1*10., y1)
# nbins=auto表示根据轴的长度自动计算可接受多少刻度，会考虑字体大小但会忽略字符串长度，如图3所示
# steps=[1, 4, 10]biao'shi表示可接受该列表倍数的刻度，eg.10 40 100,0.1 0.4 1.0
formatter = matplotlib.ticker.FormatStrFormatter('%1.1f')
locator = matplotlib.ticker.MaxNLocator(nbins='auto', steps=[1, 4, 10])
axs[1].xaxis.set_major_locator(locator)
axs[1].xaxis.set_major_formatter(formatter)

formatter = matplotlib.ticker.FormatStrFormatter('%1.5f')
locator = matplotlib.ticker.AutoLocator()
axs[2].xaxis.set_major_formatter(formatter)
axs[2].xaxis.set_major_locator(locator)

formatter = matplotlib.ticker.FormatStrFormatter('%1.5f')
locator = matplotlib.ticker.MaxNLocator(nbins=4)
axs[3].xaxis.set_major_formatter(formatter)
axs[3].xaxis.set_major_locator(locator)

plt.show()
```

```python
# 也可以为formatter指定函数，但感觉这种不是很方便
def formatoddticks(x, pos):
    """Format odd tick positions
    """
    if x % 2:
        return '%1.2f' % x
    else:
        return ''

fig, ax = plt.subplots(figsize=(5, 3), tight_layout=True)
ax.plot(x1, y1)
formatter = matplotlib.ticker.FuncFormatter(formatoddticks)
locator = matplotlib.ticker.MaxNLocator(nbins=6)
ax.xaxis.set_major_formatter(formatter)
ax.xaxis.set_major_locator(locator)

plt.show()
```

#### 日期刻度

```python
import datetime
x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
fig, ax = plt.subplots(figsize=(5, 3), tight_layout=True)
base = datetime.datetime(2017, 1, 1, 0, 0, 1)
time = [base + datetime.timedelta(days=x) for x in range(len(y1))]

ax.plot(time, y1)
ax.tick_params(axis='x', rotation=70)
plt.show()
```

```python
# 最简单的解决方法是将格式传递给matplotlib.dates.DateFormatter
# 另请注意，29日和下个月非常接近，我们可以通过使用dates.DayLocator类来解决这个问题， 该类允许我们指定要使用月份的日期列表
import matplotlib.dates as mdates
locator = mdates.DayLocator(bymonthday=[1, 15])
formatter = mdates.DateFormatter('%b %d')

fig, ax = plt.subplots(figsize=(5, 3), tight_layout=True)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
ax.plot(time, y1)
ax.tick_params(axis='x', rotation=70)
plt.show()
```
