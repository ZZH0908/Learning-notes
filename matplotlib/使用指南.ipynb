matplotlib库的内容是按照层级结构划分的，最顶层的结构为"state-machine environment" ，其通过matplotlib.pyplot模块来提供，随后使用一些简单的函数对当前图中的当前axes 进行绘图等操作；

```python
# 创建图Figure
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()  # an empty figure with no axes
fig.suptitle('No axes on this figure')  # Add a title so we know which it is

fig, ax_lst = plt.subplots(2, 2)  # Create a figure and a set of subplots.;返回一个图、一个或多个axes.Axes对象
plt.show()
```

```python
# 绘制函数的输入类型是np.array，对于其他类型数据在绘制之前要进行类型变换
b = np.matrix([[1,2],[3,4]])
b_asarray = np.asarray(b)
```

```python
# Pyplot为底层面向对象的绘图库,状态机隐式地自动创建图形和轴
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear') # 第一次调用就创建必要的图和轴
plt.plot(x, x**2, label='quadratic') # 随后的调用将在当前图当前轴绘制额外的线条
plt.plot(x, x**3, label='cubic')
# 设置label、标题、legend等都是用当前图和当前轴上操作的
plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend(loc='upper left')

plt.show()
```

```python
# 还可以把相关代码封装到一个函数中，这样就可以对不同的数据进行绘图而不用重复的写代码
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict) # 直接使用axes.Axes对象进行绘制内容，更为灵活
    return out
fig, (ax1, ax2) = plt.subplots(1, 2)
data1, data2, data3, data4 = np.random.randn(4, 100)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
plt.show()
```

```python
# 在交互式模式下
# plt.plot(x)直接显示图像，不需要调用show()函数
# 在非交互式模式下，需要在show之前需要使用plt.ioff()来关闭交互模式，不然图像会一闪而过
plt.ion()
plt.plot([1.6, 2.7])
# 也可以使用plt.gca()和plt.gcf()来获取当前轴和当前图进行修改更新
ax = plt.gca()
ax.plot([3.1, 2.2])
```

```python
for i in range(3):
    plt.plot(np.random.rand(10)) # 交互式模式下则连续绘制了三次，不用调用show也没有暂停关闭操作
```

```python
# 非交互模式下需要plt.show()后才能显示图像，而且要在手动关闭显示窗口后才继续执行
# 代码
plt.ioff()
for i in range(3):
    plt.plot(np.random.rand(10))
    plt.show() # 在脚本中运行才能看到区别，三个图意思是show执行了三次
```
