```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读取数据
f = open('testSet.txt', 'r')
datalist = f.readlines()
datasets = []

for line in datalist:
    line = line.split()
    line = [float(i) for i in line]
    datasets.append([1.0, line[0], line[1], line[2]])
datasets = np.asarray(datasets)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(datasets[:, 1], datasets[:, 2], c=15.0*datasets[:, 3])
# plt.show()
```

```python
# 梯度上升(极大似然)、算法训练
def sigmoid_fn(x):
    return 1.0/(1 + np.exp(-x))

def gradAscent(data):
    data_ = data[:, 0:3]
    label = data[:, 3].reshape(-1, 1)
    print(label.shape)
    m, n = data.shape
    alpha = 0.001
    epochs = 500
    w = np.ones((n-1, 1))
    for i in range(epochs):
        h = sigmoid_fn(np.dot(data_, w))
        loss = label - h
        w = w + alpha * np.dot(data_.transpose(), loss)
    return w

weights = gradAscent(datasets)
print(weights.shape)
# 可视化出分界线，也就是h=0.5，即w*b=0
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(datasets[:, 1], datasets[:, 2], c=15.0*datasets[:, 3])
x = np.arange(-3.0, 3.0, 0.1)
y = (-weights[0]-weights[1]*x)/weights[2]
plt.plot(x , y)
plt.show()
```

```python
# 随机梯度上升
# 计算量减少，但效果不好
def stocGradAscent0(data):
    data_ = data[:, 0:3]
    label = data[:, 3]
    print(label.shape)
    m, n = data_.shape
    alpha = 0.01
    w = np.ones(n)   # initialize to all ones
    for i in range(m):
        h = sigmoid_fn(np.sum(data_[i] * w))
        error = label[i] - h
        w = w + alpha * error * data_[i]
    return w

weights = stocGradAscent0(datasets)
print(weights)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(datasets[:, 1], datasets[:, 2], c=15.0*datasets[:, 3])
x = np.arange(-3.0, 3.0, 0.1)
y = (-weights[0]-weights[1]*x)/weights[2]
plt.plot(x, y)
plt.show()
```

```python
# 保存并绘制参数随迭代的变化
def stocGradAscent0(data):
    data_ = data[:, 0:3]
    label = data[:, 3]
    print(label.shape)
    m, n = data_.shape
    alpha = 0.01
    w = np.ones(n)   # initialize to all ones
    record_w = []
    for epoch in range(200): # 为了和原文保持同样的迭代次数，同时这也改善了分类效果，但参数收敛的很慢
        for i in range(m):
            h = sigmoid_fn(np.sum(data_[i] * w))
            error = label[i] - h
            w = w + alpha * error * data_[i]
            record_w.append(w)
    return w, np.asarray(record_w)

weights, record_w = stocGradAscent0(datasets)
print(weights)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(datasets[:, 1], datasets[:, 2], c=15.0*datasets[:, 3])
x = np.arange(-3.0, 3.0, 0.1)
y = (-weights[0]-weights[1]*x)/weights[2]
plt.plot(x, y)
plt.show()
```

```python
# 改进：采用学习率下降以及随机梯度更新
def stocGradAscent0(data):
    data_ = data[:, 0:3]
    label = data[:, 3]
    print(label.shape)
    m, n = data_.shape
    # alpha = 0.01
    w = np.ones(n)   # initialize to all ones
    record_w = []
    for epoch in range(200):
        dataIndex = list(range(m))
        for i in range(m):
            alpha = 4 / (1.0 + epoch + i) + 0.0001  
            randIndex = int(np.random.uniform(0, len(dataIndex)))  
            h = sigmoid_fn(np.sum(data_[randIndex] * w))
            error = label[i] - h
            w = w + alpha * error * data_[randIndex]
            record_w.append(w)
            del(dataIndex[randIndex])
    return w, np.asarray(record_w)


weights, record_w = stocGradAscent0(datasets)
print(weights)
max_value = record_w.max(axis=0)
min_value = record_w.min(axis=0)
print(max_value)
print(min_value)
plt.subplot(311)
plt.plot(record_w[:, 0])
plt.ylim(min_value[0], max_value[0])
plt.xlim(0, 20100)
plt.subplot(312)
plt.plot(record_w[:, 1])
plt.ylim(min_value[1], max_value[1])
plt.xlim(0, 20100)
axis3 = plt.subplot(313)
plt.plot(record_w[:, 2])
plt.ylim(min_value[2], max_value[2])
plt.xlim(0, 20100)
plt.tight_layout()
plt.show() # 很快就收敛了，稳定了，效果明显
# 分类效果就不plot了，和第一个效果差不多
```
