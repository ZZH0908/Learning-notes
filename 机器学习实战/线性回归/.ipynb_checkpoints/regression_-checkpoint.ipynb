### 线性回归是利用数理统计中回归分析，来确定两种或两种以上变量间相互依赖的定量关系的一种统计分析方法，即对连续型的数据进行预测.

### 用线性回归找到最佳拟合直线

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_data(filename):
    x_data = []
    y_data = []
    f = open(filename)
    for line in f.readlines():
        tolist = line.strip().split('\t')
        str2num = [float(i) for i in tolist]
        x_data.append([str2num[0], str2num[1]])  # str2num[0]为1.0也就是偏置
        y_data.append(str2num[-1])

    return x_data, y_data


x, y = read_data('ex0.txt')
x = np.asarray(x)
y = np.asarray(y)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x[:, 1], y)
plt.show()
```

```python
def regress(x, y):
    xTx = np.matmul(x.T, x)
    # 这里解释一下线性代数的知识，行列式为0则不是满秩，矩阵即线性变换，不满秩意味着该线性变换后降维了，降维的空间是不存在逆变换的
    if np.linalg.det(xTx) == 0:
        print("该矩阵为奇异矩阵，不存在逆矩阵~")  
    ws = np.matmul(xTx.I, np.matmul(x.T, y.T))
    return ws


x, y = read_data('ex0.txt')
x_mat = np.matrix(x)
y_mat = np.matrix(y)
w = regress(x_mat, y_mat)
print(w)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x_mat[:, 1].A, y_mat.A)
x_sort = np.sort(x_mat, axis=0)
y_ = np.matmul(w.T, x_sort.T).reshape((-1, 1))
ax.plot(x_sort[:, 1], y_, c='r')
plt.show()
```

```python
# 平均回归预测模型的好坏，即计算预测序列y_和真实y的匹配程度，相似程度，关系距离~~，求相关系数
x, y = read_data('ex0.txt')
x_mat = np.matrix(x)
y_mat = np.matrix(y)
w = regress(x_mat, y_mat)
y_ = np.matmul(w.T, x_mat.T).reshape((-1, 1))
print(np.corrcoef(y_.T, y))
```

### 局部加权线性回归

```python
# 对测试点附近的点赋更大的权重值，因为距离远的data对当前的预测影响很小
def lwlr(testPoint,xArr,yArr,k=1.0):
    m = np.shape(xArr)[0]
    # 对角矩阵用于对数据进行权重赋值
    weights = np.matrix(np.eye((m)))
    for j in range(m):                  
        diffMat = testPoint - xArr[j,:]     
        weights[j,j] = np.exp(np.matmul(diffMat, diffMat.T)/(-2.0*k**2))
    xTx = np.matmul(xArr.T, np.matmul(weights, xArr))
    if np.linalg.det(xTx) == 0.0:
        print("This matrix is singular, cannot do inverse")
        return
    ws = np.matmul(xTx.I, np.matmul(xArr.T, np.matmul(weights, yArr.T)))
    return testPoint * ws

# 对所有的点进行遍历
def lwlrTest(testArr,xArr,yArr,k=1.0):  
    m = np.shape(testArr)[0]
    yHat = np.zeros(m)
    for i in range(m):
        yHat[i] = lwlr(testArr[i],xArr,yArr,k)
    return yHat


x, y = read_data('ex0.txt')
x_mat = np.matrix(x)
y_mat = np.matrix(y)
y_ = lwlrTest(x_mat, x_mat, y_mat, k=0.05)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x_mat[:, 1].A, y_mat.A)
idex = x_mat[:, 1].argsort(0).reshape((1, -1))
x_sort = x_mat[idex][0, :, :]
ax.plot(x_sort[:, 1], y_[idex].T, c='r')
plt.show() # 欠拟合
```

```python
x, y = read_data('ex0.txt')
x_mat = np.matrix(x)
y_mat = np.matrix(y)
y_ = lwlrTest(x_mat, x_mat, y_mat, k=0.003)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x_mat[:, 1].A, y_mat.A)
idex = x_mat[:, 1].argsort(0).reshape((1, -1))
x_sort = x_mat[idex][0, :, :]
ax.plot(x_sort[:, 1], y_[idex].T, c='r')
plt.show() # 过拟合
```

```python
x, y = read_data('ex0.txt')
x_mat = np.matrix(x)
y_mat = np.matrix(y)
y_ = lwlrTest(x_mat, x_mat, y_mat, k=0.01)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x_mat[:, 1].A, y_mat.A)
idex = x_mat[:, 1].argsort(0).reshape((1, -1))
x_sort = x_mat[idex][0, :, :]
ax.plot(x_sort[:, 1], y_[idex].T, c='r')
plt.show() # 效果刚好
```

书本上还描述了不同的k值（0.1， 1，10）下的训练预测误差和测试预测误差，结果表明k值并不是越小越好，进一步比较了线性回归和局部加权线性回归，取得了相似的误差，最终
结论：1、比较效果好坏时要在测试集上进行比较才能反应真实的情况；2、k值得好坏要进行大量得实验对比得出；3、个人认为局部加权还是要比简单得线性回归能力要强，这取决于
训练集得质量数量以及k得最佳取值；

### 岭回归

岭回归就是在矩阵$x^{\tau }x$加上$\lambda I$使得矩阵变成非奇异矩阵，岭回归最初时用于处理样本特征数大于样本数的情况，但现在也用于正则化的作用对参数进行惩罚，增强泛化性

```python
def read_abalone(filename):
    x_data = []
    y_data = []
    f = open(filename)
    for line in f.readlines():
        tolist = line.strip().split('\t')
        str2num = [float(i) for i in tolist]
        line_list = []
        for i in range(8):
            line_list.append(str2num[i])
        x_data.append(line_list)  # str2num[0]为1.0也就是偏置
        y_data.append(str2num[-1])

    return x_data, y_data

def ridgeregres(x, y, lam=0.2):
    xTx_plus = np.matmul(x.T, x) + lam*np.eye(x.shape[1])
    if np.linalg.det(xTx_plus) == 0:         # 为防止lamda为0地情况，还是要进行if条件判断
        print("该矩阵为奇异矩阵，不存在逆矩阵~")
    ws = np.dot(xTx_plus.I, np.dot(x.T, y))

    return ws

def ridgetest(xmat, ymat):
    ymean = np.mean(ymat.T, 0)
    print(ymean.shape)
    y = ymat.T - ymean
    xmean = np.mean(xmat, 0)
    xvar = np.var(xmat, 0)
    x = (xmat - xmean)/xvar
    wmat = np.zeros((30, 8))
    lam_num = 30
    for j in range(lam_num):
        w = ridgeregres(x, y, np.exp(j-10))
        wmat[j, :] = w.T
    return wmat


x_data, y_data = read_abalone("abalone.txt")
x_data = np.mat(x_data)
y_data = np.mat(y_data)
w_array = ridgetest(x_data, y_data)
print(w_array.shape)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('log(lamda)')
ax.set_ylabel('weights')
ax.plot(w_array)
plt.show() # 随着lamda取值越大，W——>0
```

### 前向逐步回归

```python
它属于一种贪心算法，先设置初始权重，然后让权重减去或加上一个比较小地数，判断选择保证每一步都尽可能地减少误差
```

```python
def read_abalone(filename):
    x_data = []
    y_data = []
    f = open(filename)
    for line in f.readlines():
        tolist = line.strip().split('\t')
        str2num = [float(i) for i in tolist]
        line_list = []
        for i in range(8):
            line_list.append(str2num[i])
        x_data.append(line_list)  # str2num[0]为1.0也就是偏置
        y_data.append(str2num[-1])

    return x_data, y_data

def stageWise(xmat, ymat, eps=0.001, steps=5000):
    ymean = np.mean(ymat.T, 0)
    y = ymat.T - ymean
    xmean = np.mean(xmat, 0)
    xvar = np.var(xmat, 0)
    x = (xmat - xmean)/xvar
    m, n = xmat.shape
    ws = np.zeros((n, 1))
    finall_ws = np.zeros((steps, n))
    wbest = ws.copy()
    for step in range(steps):
        if step%100 == 0:
            print(ws.T)
        lowesterror = np.inf
        for i in range(n):
            add_sub = [-1, 1]
            for j in add_sub:
                w_traing = ws.copy()
                # 每次增加或减少一个很小的数
                w_traing[i] += eps*j
                y_ = np.dot(x, w_traing)
                current_error = ((y.A - y_.A)**2).sum()
                if current_error < lowesterror:
                    lowesterror = current_error
                    wbest = w_traing
        ws = wbest.copy()
        finall_ws[step, :] = ws.T

    return finall_ws


x_data, y_data = read_abalone("abalone.txt")
x_data = np.mat(x_data)
y_data = np.mat(y_data)
w = stageWise(x_data, y_data)
plt.plot(w)
plt.xlabel("steps")
plt.ylabel("weights")
plt.show()
```

该算法地优点在于它可以帮助人们理解现有的模型并做出改进，当构建了一个模型后，可以运行该算法找出重要的特征，这样就有可能及时停止对那些不重要特征的收集。最后，如果用于测试，该算法每100次迭代后就可以构建出一个模型，可以使用类似于10折交叉验证的方法比较这些模型，最终选择使误差最小的模型.
